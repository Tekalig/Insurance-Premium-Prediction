```python
from src.statistical_modeling import prepare_data, build_models, plot_feature_importance, evaluate_models
from src.data_processing import load_data
```

```python
# Load the dataset
filepath = '../data/MachineLearningRating_v3.txt'
data = load_data(filepath)
```

```python
# Prepare the data
X_train, X_test, y_train, y_test = prepare_data(data)
```

```python
# Build and evaluate models
model_results, models = build_models(X_train, X_test, y_train, y_test)
```

```python
# Print evaluation metrics
evaluate_models(model_results)
```

```python
# Feature importance for Random Forest model
for model_name, model in models.items():
    model.fit(X_train, y_train)
    plot_feature_importance(model, X_train)
    print('__'*25)
    print('#*'*25)
```
